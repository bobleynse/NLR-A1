```python
import matplotlib.pyplot as plt
```

```python
# Different amount of training steps
# all with 400000 training lines
# all with the initial [100, 150, 100, 50] model
# 5 train 2 test

steps = [0, 500, 1000, 1500]
accuracy = [0, 0.704, 0, 0]
plt.scatter(steps, accuracy)
plt.title('Training accuracy based on different step-sizes')
plt.xlabel('Training steps')
plt.ylabel('Training accuracy')
plt.show()
```

```python
# Different rectsizes
# all with 200000 training lines
# all with the initial [100, 150, 100, 50] model
# 5 train 2 test

steps = [0, 3, 5, 7]
accuracy = [0, 0.69, 0.66, 0.705]
plt.scatter(steps, accuracy)
plt.title('Training accuracy based on different rect-sizes')
plt.xlabel('Rect size')
plt.ylabel('Training accuracy')
plt.show()
```
