```python
from osgeo import gdal, ogr

src_ds = gdal.Open( "zh1.tif" )

data = src_ds.ReadAsArray()
```

```python
print(data)
```

```python
plt.imshow(data[3])
plt.show()
```

```python
np.amax(data[3])   
```

```python
# size
x = 0
y = 400

data_comp = [data[1][x:y, x:y], data[2][x:y, x:y], data[3][x:y, x:y]]
```

```python



output_list = []

for row in range(len(data_comp[0])):
    output_list.append([])
    for col in range(len(data_comp[0][row])):
        output_list[row].append([])
        for band in range(3):
            output_list[row][col].append(data_comp[band][row][col] / 1200)
            
print(np.array(output_list))
```

```python
plt.imshow(output_list)
plt.show()
```
